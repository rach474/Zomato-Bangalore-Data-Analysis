### <b>Let us first define what information the columns contain based on our inspection</b>

```python
#Dropping the columns which are not useful for data analysis

data=data.drop(['url','address','phone','dish_liked','reviews_list','menu_item'],axis=1)
data.head()
```

```python
data.info()
```

### <b>There are a lot of null values in rate column,and few null values in location,rest_type,cuisines and approx_cost(for two people) that needs to be handled.</b>

# Dropping Duplicates

```python
data.drop_duplicates(inplace=True)

# Shape of the updated dataframe
data.shape
```

### We have dropped the duplicate rows

### <b>Let us try to clean and handle these NaN values one column at a time</b>

# Cleaning Rate Column

```python
data['rate'].unique()
```

# Removing "NEW", "-" and "/5" from Rate Column

```python
#We will Replace "NEW" and "-" values with NaN values and split the rest of the values at "/"

def handlerate(value):
    if(value=='NEW' or value=='-'):
        return np.nan
    else:
        value=str(value).split('/')
        value=value[0]
        return float(value)

data['rate']=data['rate'].apply(handlerate)
data['rate'].unique()
```

```python
data.rate.isnull().sum()
```

### <b>There are 10019 null values present in rate column</b>

# Filling Null Values in Rate Column with Mean

```python
data['rate'].fillna(data['rate'].mean(),inplace=True)
data['rate'].isnull().sum()
```

```python
data['rate'].unique()
```

### <b>Lets find the percentage of rows that contain atleast one NaN value</b>

```python
#Percentage of rows containing one NaN value
total_nan = sum([True for id,row in data.iterrows() if any(row.isnull())])

perc_nan = round(total_nan/51609*100, 2)

print(f'Hence, there are a total of {total_nan}({perc_nan}%) rows which contain atleast one NaN value.')
```

### <b>Since approximately 1.1% of all the rows in the dataset contains atleast one NaN value, it is practical to eliminate all the rows containing NaN values. Eliminating NaN values won't hamper our analysis.</b>

# Dropping Null Values

```python
data.dropna(inplace=True)

#Checking for null values in every column
data.isnull().sum()
```

Hence, no null values are present in the dataset

```python
data.head()
```

```python
#Changing the names of some columns to more appropriate names

data.rename(columns={'approx_cost(for two people)':'Cost2plates','listed_in(type)':'type','listed_in(city)':'city'},inplace=True)
data.head()
```

```python
data['location'].unique()
```

```python
data['city'].unique()
```

# <b>location and city-both columns give the same type of information. So we can remove either location or city in our dataset.<b>

```python
data=data.drop(['city'],axis=1)
data.head()
```

# Cleaning Cost2plates Column

```python
data['Cost2plates'].unique()
```

# Removing ',' from Cost2Plates Column

```python
#Removing commas from the values of Cost2plates column and converting them to float data type

def handlecomma(value):
    value=str(value)
    if ',' in value:
        value=value.replace(',','')
        return float(value)
    else:
        return float(value)
    
data['Cost2plates']=data['Cost2plates'].apply(handlecomma)
data['Cost2plates'].unique()
```

```python
data.head()
```

# Cleaning rest_type Column

```python
# Finding the different values the 'rest_type' column takes

data['rest_type'].value_counts()
```

### <b>We will make a cluster of the values having fewer counts and rename it as 'others'.</b>

```python
rest_types=data['rest_type'].value_counts(ascending= False)
rest_types
```

```python
# Making a cluster of all the values having counts less than 1000

rest_types_lessthan1000=rest_types[rest_types<1000]
rest_types_lessthan1000
```

# Making rest_types_lessthan100 in frequency as 'others'

```python
def handle_rest_type(value):
    if value in rest_types_lessthan1000:
        return 'others'
    else:
        return value
    
data['rest_type']=data['rest_type'].apply(handle_rest_type)
data['rest_type'].value_counts()
```

```python
data.head()
```

# Cleaning Location Column

```python
data['location'].value_counts()
```

```python
# Making a cluster of all the values having counts less than 300 and rename it as 'others'

location=data['location'].value_counts(ascending=False)
location_lessthan300=location[location<300]

# Making location_lessthan300 in frequency as 'others'

def handle_location(value):
    if value in location_lessthan300:
        return 'others'
    else:
        return value
    
data['location']=data['location'].apply(handle_location)
data['location'].value_counts()
```

```python
data.head()
```

# Cleaning Cuisines Column

```python
data['cuisines'].value_counts()
```

```python
# Making a cluster of all the values having counts less than 100 and rename it as 'others'

cuisines=data['cuisines'].value_counts(ascending=False)
cuisines_lessthan100=cuisines[cuisines<100]

# Making cuisines_lessthan100 in frequency as 'others'

def handle_cuisines(value):
    if value in cuisines_lessthan100:
        return 'others'
    else:
        return value
    
data['cuisines']=data['cuisines'].apply(handle_cuisines)
data['cuisines'].value_counts()
```

```python
data.head()
```

```python
data['type'].value_counts()
```

### Type column has few unique values. Therefore it is already clean
